### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
numericVal = df.select_dtypes(include = ['number'])
for col in numericVal:
    meanVal = numericVal[col].mean()
    standardVal = numericVal[col].std()
    maxVal = numericVal[col].max()
    minVal = numericVal[col].min()
    modeVal = numericVal[col].mode()

for col in numericVal :
    print(col)
    print("Average :",meanVal)
    print("Standard Deviation :",standardVal)
    print("Maximum :",maxVal)
    print("Minimum :",minVal)
    print("Range :",minVal ,"to", maxVal)
```

### 6)  For the qualitative attribute (class), count the frequency for each of its distinct values.



```python
frequency = df['Pclass'].value_counts()
print(frequency)
```

###  7) It is also possible to display the summary for all the attributes simultaneously in a table using the describe() function. If an attribute is quantitative, it will display its mean, standard deviation and various quantiles (including minimum, median, and maximum) values. If an attribute is qualitative, it will display its number of unique values and the top (most frequent) values.

```python
print("Qualitative")
print(df.describe(include='object') )
print("Quantative")
print(df.describe() )

```

### 8)  For multivariate statistics, you can compute the covariance and correlation between pairs of attributes.



```python
numberVal = df.select_dtypes(include = ['number'])
covariance = numberVal.cov()
print("Covarience:")
print(covariance)

correlation = numberVal.corr()
print("Correlation:")
print(correlation)
```

### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
age_data = df['Age'].dropna()

plt.hist(age_data, bins=8, edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age Range')
plt.ylabel('Frequency')

plt.show()
```

### 10) A boxplot can also be used to show the distribution of values for each attribute.

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df = numeric_df.dropna()
plt.figure(figsize=(12, 6))
plt.boxplot(numeric_df.values, tick_labels=numeric_df.columns, patch_artist=True)
plt.title('Boxplot of All Numeric Attributes')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
df = df.dropna(subset=['Age', 'Fare', 'SibSp', 'Parch', 'Pclass'])
pairs = [
    ('Age', 'Fare'),
    ('Age', 'Pclass'),
    ('Fare', 'Pclass'),
    ('SibSp', 'Parch'),
    ('Age', 'SibSp')
]

plt.figure(figsize=(15, 10))

for i in range(len(pairs)):
    x = pairs[i][0]
    y = pairs[i][1]
    plt.subplot(2, 3, i+1)
    plt.scatter(df[x], df[y], alpha=0.6)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'{x} vs {y}')

plt.tight_layout()
plt.suptitle('Scatter Plots of 5 Attribute Pairs', fontsize=16, y=1.02)
plt.show()
```
